#### 4.1. Write a Python function that calculates the factorial of a given number.


```python
def factorial(n):
    result = 1
    for i in range(1,n+1):
        result *= i
    return result
```

#### 4.2. Test your function with different input values and observe the results.

#### 5.1. Understand what a kernel is in the context of Jupyter Notebooks.

#### 6.3. Discuss when different export formats might be useful.

Jupyter Notebooks provide various export formats to cater to different needs and use cases. The choice of export format depends on the audience, purpose, and the tools/platforms used for further analysis or presentation. Here's a discussion on when different Jupyter Notebook export formats might be useful:

**1. HTML:**

`Use Case:` Web Sharing, Documentation

`Why:` Exporting to HTML is useful when you want to share your Jupyter Notebook as a static web page. It preserves the structure, code, and visualizations, making it suitable for documentation or sharing interactive content online.

**2. PDF:**

`Use Case:` Formal Reports, Printing

`Why:` PDF is suitable for creating a static, printable version of your Jupyter Notebook. It's often used for formal reports, presentations, or when you need a document that looks consistent across different platforms.

**3. Markdown:**

`Use Case:` Documentation, Version Control

`Why:` Exporting to Markdown is helpful when you want to convert your Jupyter Notebook into a plain text format that retains basic formatting. This is often used for documentation and is compatible with version control systems like Git.

**4. Notebook (.ipynb):**

`Use Case:` Collaboration, Future Editing

`Why:` Keeping the notebook in its native format (.ipynb) allows for easy collaboration with others who use Jupyter Notebooks. It also enables you to continue working on the notebook in the future, preserving code cells, outputs, and interactivity.

**5. Slides (reveal.js, Slides, ...)**

`Use Case:` Presentations

`Why:` Jupyter Notebooks can be converted into interactive slides using various slide show formats like reveal.js. This is useful for presenting your work during meetings or conferences, especially when the notebook contains both code and visualizations.

**Executable Script (Python, R, ...):**

`Use Case:` Code Execution Outside Jupyter

`Why:` Exporting the notebook as an executable script in Python (.py) or another language format is useful when you want to run the code outside of the Jupyter environment. This can be beneficial for integration into a larger codebase or automation workflows.

**Latex:**

`Use Case:` Academic Writing, Publishing

`Why:` Exporting to LaTeX is valuable when you are involved in academic writing or publishing, as it allows you to incorporate Jupyter Notebook content into LaTeX documents seamlessly. This is especially useful for creating mathematical and scientific documents.

**Custom Export Formats:**

`Use Case:` Specialized Needs

`Why:` Jupyter Notebooks allow the creation of custom export formats through templates. This is useful when you have specific styling or formatting requirements that are not met by the standard export options.

> When choosing an export format, consider the intended use of the exported content and the preferences of your audience. Different formats serve different purposes, and selecting the right one enhances the effectiveness of sharing and presenting your work.

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_func(w=1):
    x = np.random.uniform(0,5, size = 100)
    y = 2 * x + w * np.random.normal(size=100)
    plt.scatter(x, y)
```

```python
widgets.interact(plot_func, w=(0, 5, 0.5))
```
